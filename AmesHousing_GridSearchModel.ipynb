```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Ames_Housing_Fina_Data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
X=df.drop('SalePrice',axis=1)
y=df['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)

# Scale Data

from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
scaler.fit(X_train)

X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import ElasticNet
```

```python
elastic_net_model=ElasticNet(max_iter=1000000)
```

```python
param_grid={'alpha':[.1,1,5,10,100],'l1_ratio':[.1,.7,.99,1]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_model=GridSearchCV(estimator=elastic_net_model,param_grid=param_grid,
                       scoring='neg_mean_squared_error',
                       cv=5,verbose=2)
```

```python
grid_model.fit(X_train,y_train)
```

```python
grid_model.best_estimator_
```

```python
grid_model.best_params_
```

```python
pd.DataFrame(grid_model.cv_results_)
```

```python
y_pred=grid_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
mean_absolute_error(y_test,y_pred)
```
